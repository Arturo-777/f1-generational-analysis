## Este notebook realiza un an√°lisis por generaciones de pilotos bas√°ndose en su a√±o de debut y Win Rate (tasa de victorias).

## Se construye una m√©trica compuesta IFG (Indicador de Fortaleza Generacional) y se analiza la dispersi√≥n y el volumen de pilotos por generaci√≥n.

### 1Ô∏è‚É£ Cargar Datos desde la Vista SQL

```python
import pyodbc
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Crear conexi√≥n a la base de datos SQL Server
conn = pyodbc.connect(
    'DRIVER={ODBC Driver 17 for SQL Server};'
    'SERVER=localhost;'
    'DATABASE=F1 Drivers;'
    'Trusted_Connection=yes;'
)

# Leer datos desde la vista
query = "SELECT * FROM dbo.vw_F1_Generation_Performance;"
df = pd.read_sql(query, conn)

# Vista general del dataset
df.head()
```

### 2Ô∏è‚É£ Preparaci√≥n de Datos ‚Äî A√±os de Debut por Piloto

```python
# Cada piloto tiene m√∫ltiples filas (una por temporada),se debe eliminar la duplicidad.
# Se agrupa por piloto para determinar su a√±o de debut y su Win Rate.

df_debut = (
    df.groupby("Driver", as_index=False)
    .agg({
        "Season": "min",      # A√±o m√≠nimo de debut
        "Win_Rate": "first"   # Win Rate √∫nico del piloto
    })
    .rename(columns={"Season": "debut_year", "Win_Rate": "win_rate"})
)

df_debut.head()
```

### 3Ô∏è‚É£ Clasificaci√≥n de Generaciones

```python
# Se asigna una generaci√≥n con base en el a√±o de debut del piloto y a la decada que pertenece.
# Las generaciones seleccionadas son: (1990, 2000, 2010, 2020).

def asignar_generacion(year):
    if 1990 <= year < 2000:
        return 1990
    elif 2000 <= year < 2010:
        return 2000
    elif 2010 <= year < 2020:
        return 2010
    elif year >= 2020:
        return 2020
    else:
        return np.nan

df_debut["Generation"] = df_debut["debut_year"].apply(asignar_generacion)
df_debut.head()
```

### 4Ô∏è‚É£ C√°lculo de M√©tricas por Generaci√≥n

```python
# Se calculan las m√©tricas base:
# - mean_winrate : Promedio del Win Rate (tasa de victoria) por generaci√≥n
# - std_winrate  : Desviaci√≥n est√°ndar del Win Rate
# - count_pilots : N√∫mero de pilotos en esa generaci√≥n
# - IFG          : Indicador de Fortaleza Generacional *Metrica clave*
# - CV           : Coeficiente de variaci√≥n (dispersi√≥n relativa)



# Generaciones de inter√©s
generations = [1990, 2000, 2010, 2020]

# Filtrar pilotos pertenecientes a las generaciones seleccionadas
df_tmp = df_debut[df_debut["Generation"].isin(generations)].copy()

# Agrupar por generaci√≥n
grouped = df_tmp.groupby("Generation")

# C√°lculo individual de m√©tricas (compatible con versiones antiguas de pandas)
mean_winrate = grouped["win_rate"].mean()
std_winrate  = grouped["win_rate"].std()
count_pilots = grouped["Driver"].nunique()

# Construir DataFrame final de m√©tricas
metricas_df = pd.DataFrame({
    "Generation": mean_winrate.index,
    "mean_winrate": mean_winrate.values,
    "std_winrate": std_winrate.values,
    "count_pilots": count_pilots.values
}).reset_index(drop=True)

# Calcular IFG y Coeficiente de Variaci√≥n (CV)
metricas_df["IFG"] = metricas_df["mean_winrate"] * np.log1p(metricas_df["count_pilots"])
metricas_df["CV"] = np.where(
    metricas_df["mean_winrate"] != 0,
    metricas_df["std_winrate"] / metricas_df["mean_winrate"],
    np.nan
)

# Ordenar cronol√≥gicamente
metricas_df = metricas_df.sort_values("Generation").reset_index(drop=True)

# Mostrar resultados tabulares
print("üìä M√©tricas por Generaci√≥n (con IFG y CV):")
display(metricas_df)


```

### 5Ô∏è‚É£ Visualizaci√≥n de Resultados

```python
# Gr√°fico combinado:
# - Barras azules: cantidad de pilotos por generaci√≥n
# - L√≠nea roja: m√©trica IFG (fortaleza generacional)

# Asegurar que 'Generation' sea string para eje X
metricas_df["Generation"] = metricas_df["Generation"].astype(int).astype(str)

# Crear figura
fig, ax1 = plt.subplots(figsize=(9,6))

# --- Barras: cantidad de pilotos ---
sns.barplot(
    x="Generation", y="count_pilots", data=metricas_df,
    ax=ax1, color="skyblue", zorder=1
)
ax1.set_ylabel("Cantidad de Pilotos", fontsize=11)
ax1.set_xlabel("Generaci√≥n", fontsize=11)
ax1.set_ylim(0, metricas_df["count_pilots"].max() * 1.3)

# Etiquetas sobre las barras
for i, v in enumerate(metricas_df["count_pilots"]):
    ax1.text(i, v + (metricas_df["count_pilots"].max() * 0.03), str(int(v)), 
             ha='center', va='bottom', fontsize=10)

# --- L√≠nea: IFG ---
ax2 = ax1.twinx()
sns.lineplot(
    x="Generation", y="IFG", data=metricas_df,
    ax=ax2, color="red", marker="o", linewidth=2, zorder=2
)
ax2.set_ylabel("IFG (Indicador de Fortaleza Generacional)", fontsize=11)
ax2.set_ylim(0, metricas_df["IFG"].max() * 1.3)

# --- Est√©tica final ---
ax1.grid(axis="y", linestyle="--", alpha=0.7, zorder=0)
plt.title("Fortaleza Generacional en F1 ‚Äî Win Rate, Consistencia y Volumen de Pilotos", fontsize=14)
plt.tight_layout()
plt.show()
```
